Line Plot 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("iris")
```

```python
for col in df.columns:
    print(f'{col} : {df[col].unique()}')
```

```python
sns.set_style('darkgrid')
```

```python
setosa_df=df[df.species=='setosa']
versicolor_df=df[df.species=='versicolor']
virginica_df=df[df.species=='virginica']
```

```python
plt.figure(figsize=(10,7))
plt.plot(df.sepal_width, df.sepal_length);
```

```python
import matplotlib as mtp
mtp.rcParams['figure.figsize']=(10,7)
sns.scatterplot(x=df.sepal_width,y=df.sepal_length);
```

```python
plt.plot(df.sepal_width,df.sepal_length, '4-.r', markersize=12, alpha=.5 )
```

Histogram

```python
import numpy as np
plt.hist([df.sepal_width, df.sepal_length], bins=np.arange(2,10,0.5), stacked=True);
```

Barplot

```python
years=range(2000,2006)
apples=[0.35,0.6,0.9,0.8,0.65,0.8]
oranges=[0.4,0.8,0.9,0.7,0.6,0.8]
```

```python
plt.plot(years,apples,'o-.r',markersize=12)
plt.plot(years,oranges,'D-.g',markersize=8);
```

```python
''' Bar plot is same as line plot, but here we are having bars whose height is 
    based on parameter on y-axis'''
plt.plot(years,apples,'D-.r',markersize=12)
plt.bar(years,apples);
```

```python
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples)
plt.legend(['apples','oranges'])
plt.plot(years,apples,'D-.k',markersize=8);
```

Heat Map


```python
flights_df = sns.load_dataset('flights').pivot('month','year','passengers')
flights_df
```

```python
plt.title("Number of Passengers in 1000s")
sns.heatmap(flights_df)
```

```python
sns.heatmap(flights_df, fmt='d', annot=True, cmap='Blues')
plt.title("Number of Passengers in 1000s");
```

```python
fig, axes = plt.subplots(2,3, figsize=(22,12))


# Use the axes for plotting
axes[0,0].plot(years, apples, 'o--r')
axes[0,0].plot(years, oranges, 'D--b')
axes[0,0].set_xlabel('Years')
axes[0,0].set_ylabel('Yield (tonnes per hect)')
axes[0,0].set_title('Yield of Apples and Oranges')
axes[0,0].legend(['apples','orranges'])


# Pass the axes into seaborn
axes[0,1].set_title('Sepal- len vs wid')
sns.scatterplot(x=df.sepal_length,
               y=df.sepal_width,
               hue=df.species,
               s=100,
               ax=axes[0,1]
              )


# Use the axes for plotting
axes[0,2].set_title("Distrubution of Sepal Width")
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
               bins=np.arange(2,5,0.25),
               stacked=True
              )
axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);



# Pass the axes into seaborn
axes[1,0].set_title('Restaurent bills')
sns.barplot(y='day',x='total_bill',hue = 'sex', data=tips_df, ax=axes[1,0]);


# Pass the axes into seaborn
axes[1,1].set_title("Flight Traffic")
sns.heatmap(flights_df, fmt='d', annot=True, cmap='Blues', ax=axes[1,1])

# For axes[1,2] we have not set any plot

plt.tight_layout(pad=2);
```
